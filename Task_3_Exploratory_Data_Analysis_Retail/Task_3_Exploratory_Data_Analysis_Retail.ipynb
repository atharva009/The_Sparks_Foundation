### Observations

- <b>O1</b>: East & West Region has more sales compared to the other regions

```python
df_aggregated_sales = pd.DataFrame(data.groupby("Region")[['Sales','Profit']].sum()).reset_index()
df_aggregated_sales
```

```python
plt.figure(figsize=(5,5))
sns.set(style = "whitegrid", font_scale = 1.2)
sns.barplot(x = "Region", y ="Sales", data = df_aggregated_sales)
```

- <b>O2</b>: Technology products are majorly sold by this store. Therefore, so they must focus more on technology products

```python
df_aggregated_cate = pd.DataFrame({'Sales':data.groupby('Category')['Sales'].sum()}).reset_index()
df_aggregated_cate
```

```python
plt.figure(figsize = (6,6))
sns.set(style = "whitegrid")
sns.barplot(x="Category",y="Sales",data = df_aggregated_cate)
```

- <b>O3</b>: When Sales & Profit are compared, we can conclude that furniture have less profit , the reason may be stores are providing more discount which results to less profits. On the other side, Technology & Office Supplies have higher profits.

```python
la = pd.DataFrame(data.groupby('Category')[['Sales','Profit']].sum()).reset_index()
la
```

```python
w = 0.4
bar1 = np.arange(len(la['Category']))
bar2 = [i+w for i in bar1]
plt.figure(figsize=(6,6))
plt.bar(x=bar1,height='Sales',width = w,label='sales',data=la)
plt.bar(x=bar2,height='Profit',width = w,label='profit',data=la)
plt.ylabel("Amount")
plt.xlabel('Categories')
plt.xticks(bar1+w/2,la['Category'])
plt.title("Profit & Sales")
plt.legend()
```

- <b>O4</b>: In the following plots, we can observe that states like California and New York are outperforming. It is very likely that setting up new store over here would benefit the store company. But the states like Texas , Pennyslvania, Ohio , North California & more have maximum loss, so they should focus more on their marketing strategies. <br>Also, it seems like California & New York have other marketing strategies which are beneficial. While other states like Texas, Pennyslvania, Ohio, North California provide heavy discounts to attract customers but may be its not that effective. So, there is possibility that discounts can be a reason for their loss.

```python
la_states = pd.DataFrame(data.groupby('State')[['Sales','Profit']].sum()).reset_index()
la_states
```

```python
w = 0.6
bar1 = np.arange(len(la_states['State']))
bar2 = [i+w for i in bar1]
#print(bar1,bar2)
plt.figure(figsize=(18,10))
sns.set(style = "whitegrid")
plt.bar(x=bar1,height='Sales',width = w,label='sales',data=la_states)
plt.bar(x=bar2,height='Profit',width = w,label='profit',data=la_states)
plt.ylabel("Amount")
plt.xlabel('State')
plt.xticks(bar1+w/2,la_states['State'],rotation =90)
plt.title("States Performance in Profit & Sales")
plt.legend()
```

```python
plt.figure(figsize=(15,8))
sns.countplot(x=data['State'])
plt.xticks(rotation=90)
plt.show()
```

 - <b>O5</b>: Phones and Chairs have the higher sales and are making decent amount of profits. Also, the Copiers are making good amount of profits looking at their sales. On the contrary, Tables and Bookcases are facing loses.

```python
ps = data.groupby('Sub-Category')[['Sales','Profit']].sum().sort_values(by='Sales',ascending=False)
ps[:].plot.bar(color=['red','lightblue'],figsize=(15,10))
plt.title('Profit/loss & Sales across states')
plt.xlabel('Sub-Category')
plt.ylabel('Profit/loss & Sales')
plt.show()
```
